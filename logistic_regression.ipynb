```python
from urllib.request import urlretrieve
import pandas as pd
URL= "https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/7_logistic_reg/Exercise/HR_comma_sep.csv"
def load_data(url):
  urlretrieve(url, "HR.csv")
  return pd.read_csv("HR.csv")
df= load_data(URL)
df
```

```python
depts = df[['Department']]
```

```python
depts.Department.unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()

tf_depts = ohe.fit_transform(depts[['Department']])
```

```python
tf_depts = tf_depts.toarray()
```

```python
encoded_cat_df = pd.DataFrame(tf_depts, columns=ohe.get_feature_names_out(['Department']))
```

```python
encoded_cat_df
```

```python
post_ohe = pd.concat([df, encoded_cat_df], axis="columns")
```

```python
post_ohe
```

```python
post_ohe=post_ohe.drop(["Department", "Department_technical"], axis=1)
```

```python
post_ohe['salary']
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['low', 'medium', 'high']])
post_ohe['salary']= oe.fit_transform(post_ohe[['salary']])
```

```python
post_ohe
```

```python
df=post_ohe
```

```python
df.corr()[['left']]
```

```python
left= df[df.left==1]
```

```python
df.groupby('left').mean()
```

```python
df
```

```python
fdf= df[['satisfaction_level', 'salary', 'promotion_last_5years', 'average_montly_hours']]
```

```python
fdf
```

```python
labels = df[['left']]
```

```python
labels
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(fdf, labels, test_size=0.2, random_state=42)
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

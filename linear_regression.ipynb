```python
import urllib.request
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
URL = "https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/5_one_hot_encoding/Exercise/carprices.csv"
def fetch_and_load_data(URL):
    urllib.request.urlretrieve(URL, "carprices.csv")
    df = pd.read_csv("carprices.csv")
    return df
df = fetch_and_load_data(URL)
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dfle = le.fit_transform(df['Car Model'])
dfle
```

```python
df['Car Model']= dfle
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
car_cats= df[['Car Model', 'Mileage', 'Age(yrs)']]
y = df[['Sell Price($)']]

car_cats
```

```python
fitted = ohe.fit_transform(car_cats[['Car Model']])

fitted= fitted.toarray()
```

```python
fitted
```

```python
encoded_df = pd.DataFrame(fitted, columns=ohe.get_feature_names_out(['Car Model']), index=car_cats.index)
```

```python
encoded_df
```

```python
df
```

```python
cols = pd.concat([df, encoded_df], axis="columns")
```

```python
cols
```

```python
df= cols.drop(['Sell Price($)', 'Car Model_2', 'Car Model'], axis=1)
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
lin_reg= LinearRegression()

lin_reg.fit(df, cols[['Sell Price($)']])
```

```python
# lin_reg.predict([])
lin_reg.score(df, cols[['Sell Price($)']])
```
